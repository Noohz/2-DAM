## Diccionarios

### Un diccionario es una lista asociativa. Qu quiere decir esto?


* **Cada componente de un diccionario es un par clave:valor.**

* **Se delimitan por llaves { }.**

* **Los elementos se separan por comas.**

* **Son heterogneos es decir, pueden estar conformados por elementos de distintos tipo, incluidos otros diccionarios.**

* **Sus claves son inmutables y sus valores son mutables: sus elementos pueden modificarse.**


* **Las claves pueden ser valores simples de diferentes tipos y los valores asociados a la clave pueden ser datos simples o datos estructurados de tipo lista, tupla, conjunto, e inclusive otro diccionario.**

* **No se pueden modificar las claves pero, si se pueden modificar los valores que estn asociados a las claves, siempre que sean datos modificables.**

### 9. Ordenar Diccionarios

La ordenacin de los registros en los diccionarios no es tarea tan fcil como en las listas, ya que estos se pueden ordenar tanto por la clave como por el valor. 

Vamos a ver como realizar la ordenacin de diccionarios.

```python
#creamos un diccionario de clientes. La clave es su nombre y el valor lo que han gastado.
clientes = {'Pedro': 1.97,
           'Enrique': 8.49,
           'Cristian': 9.79,
           'Maria': 8.53,
           'Cristina': 0.43}
```

* **Ordenando por clave**

```python
clientes_ordenados = sorted(clientes)
print(clientes_ordenados)
```

Si se desea que el mtodo sorted() **devuelva una lista donde aparezcan el valor de cada cliente, en lugar de solamente las claves**, es necesario utilizar el mtodo **items**. 

Para ello es necesario escribir algo como:

```python
clientes_ordenados = sorted(clientes.items())
clientes_ordenados
```

* **Ordenar por Valor**

    * Si se desea ordenar el diccionario utilizando el valor en lugar de la clave es necesario **importar previamente la libreria operator.** 
    * Ahora se puede indicar la clave de ordenacin al mtodo sorted(). 
    * Para ordenar por el valor este ha de ser **operator.itemgetter(1).**
    * Para ordenar por la clave se ha de utilizar **operator.itemgetter(0).**
    
A modo de ejemplo, se pueden ordenar los clientes de mayor a menor gasto con el siguiente cdigo:

```python
import operator

clientes_ordenados = sorted(clientes.items(), key=operator.itemgetter(1), reverse=True)
clientes_ordenados
```

**Ntese que el mtodo sorted() devuelve siempre una lista de tuplas. En la tupla la clave el es primer elemento y el valor el segundo.**

### Ejercicio 3 - Numero de palabras de un texto.

Desarrolla un algoritmo que dado un texto cualquiera, nos diga el n de veces que aparece cada palabra.

```python
texto = "Una maana, tras un sueo intranquilo, Gregorio Samsa se despert convertido en un monstruoso insecto."

palabras = texto.lower().split()

DiccionarioPalabras = {}

for palabra in palabras:
    if palabra in DiccionarioPalabras:
        DiccionarioPalabras[palabra] += 1
    else:
        DiccionarioPalabras[palabra] = 1

for palabra, veces in DiccionarioPalabras.items():
    print(f"{palabra}: {veces}")
```

### Ejercicio 4.

```python
# Lo ms importante en este ejercicio es saber 
#  como iterar sobre el diccionario

# Vamos a iterar sobre los campos clave.
palabras_a_borrar = []

for palabra, veces in DiccionarioPalabras.items():
    if veces == 1:
        palabras_a_borrar.append(palabra)

# Eliminar las palabras con frecuencia 1 del diccionario
for palabra in palabras_a_borrar:
    del DiccionarioPalabras[palabra]
```

* Los elementos se separan por comas.
* Son heterogneas es decir, pueden estar conformadas por elementos de distintos tipo, incluidos otros diccionarios.
* Sus claves son inmutables y sus valores son mutables: sus elementos pueden modificarse.

**Como la siguiente sintaxis:**

* diccionario = {clave: valor, clave: valor, clave: valor, ..., clave: valor}

* Las claves pueden ser valores simples de diferentes tipos y los valores asociados a la clave pueden ser datos simples o datos estructurados de tipo lista, tupla, conjunto, e inclusive otro diccionario.
