### 1. Listas

Las listas en Python:

* Las listas son conjuntos ordenados de elementos (nmeros, cadenas, listas, etc). Las listas se delimitan por corchetes [ ].

* Los elementos se separan por comas.

* Son heterogneas. Es decir, pueden estar conformadas por elementos de distintos tipo, incluidas otras listas.

* Son mutables: sus elementos pueden modificarse.

* Python hace que este tipo de estructuras sean muy flexibles, permitiendo manipular las listas de mltiples formas.

* Permite la repeticin de elementos.


**Formas de declarar una lista:**


```python
#lista1  #lista vaca
lista1 = []

#lista2 #lista con los 10 primeros nmeros
lista2 = list(range(10))

lista3 = ['ngel', 'Fernando', 'Yasyn', 'Sal']#lista de nombres

lista4 = ["Pedro", 5, 6, "Luis", 8.5]#lista de distintos tipos
```

Ahora vamos a mostrar el contenido de cada lista.

```python
print("Lista 1:")
if lista1:#esto me dice si la lista tiene elementos
    for elemento in lista1:
        print(elemento)

print("Lista 2:")
if lista2:#esto me dice si la lista tiene elementos
    for elemento in lista2:
        print(elemento)
        
print("Lista 3:")
if lista2:#esto me dice si la lista tiene elementos
    for elemento in lista3:
        print(elemento)

print("Lista 4:")
if lista2:#esto me dice si la lista tiene elementos
    for elemento in lista4:
        print(elemento)        
```

### 1.1 Accediendo a los elementos de una lista - Slicing

Python es muy flexible en la notacin de ndices para manejo de los elementos dentro de una lista. Esta propiedad se denomina **slicing (rebanar) igual que en las strings**

Se puede acceder a los componentes de una lista mediante un **ndice entre corchetes.**


```python
#lista3 = ['ngel', 'Fernando', 'Yasyn', 'Sal']
lista3[3]
```

```python
lista3[2]
```

Puedes usar la **numeracin negativa** dentro de los ndices de la lista para obtener un elemento dentro de ella. 

Tienes que tener en cuenta que el elemento -1 es el ltimo, el -2 el penltimo, etc.

```python
#lista3 = ['ngel', 'Fernando', 'Yasyn', 'Sal']
lista3[-1] #LTIMO
```

```python
lista3[-4]
```

* **Acceder a varios elementos mediante un rango para el ndice:**

```python
#desde el elemento 1 hasta el 4 sin incluirlo
#lista2 = list(range(10))
lista2[1:4]
```

```python
#desde el principio de la lista hasta el elemento 4 sin incluirlo
#Completa el cdigo para que la salida sea
lista2[0:4]
```

```python
lista2[0:4]
```

```python
lista2 = list(range(10))
#Completa el cdigo para que la salida sea
resultado = lista2[:7]
print(resultado)
```

```python
#todos los elementos menos el ltimo
#Completa el cdigo para que la salida sea

resultado = lista2[:-1]
print(resultado)
```

```python
#elementos desde el penltimo hasta el final
#Completa el cdigo para que la salida sea
resultado = lista2[-2:]
print(resultado)
```

---
 * **Acceder mediante un intervalo entre valores, el tercer ndice**

```python
#El tercer ndice indica el incremento 
#Completa el cdigo para que la salida sea
resultado = lista2[::2]
print(resultado)
```

```python
# desde el ndice 0 hasta el final (sin contar el ltimo) de dos en dos 
#Completa el cdigo imaginando que no sabes cuntos elementos hay en lista2
resultado = lista2[0:-1:2]
print(resultado)
```

```python
# toda la lista con intervalos de 3
#si no pongo nada en el primer indice quiere decir que va desde el principio
# si no pongo nada en el segundo indice quiere decir que va hasta el final
# y si no pusiera nada en el tercer indice quiere decir que el incremento va de 1 en 1
resultado = lista2[::3]
print(resultado)
```

```python
#invierte la lista. Va desde el final al principio con un incremento de -1
resultado = lista2[::-1]
print(resultado)
```

**Ejercicio 3.**

Escribir un programa que almacene las asignaturas de un curso (por ejemplo Matemticas, Fsica, Qumica, Historia y Lengua), posteriormente, se pregunte al usuario qu nota has sacado en (asignatura) ?, guardar cada calificacin en una nueva lista, finalmente, recorrer ambas listas para ir mostrando por pantalla el nombre de la asignatura y la calificacin obtenida.

#### Solucin:

```python
# Lista de asignaturas... COMPLETA el ejercicio
asignaturas = ['Matematicas','Fisica','Quimica','Hstoria', 'Lengua']

notas = []

for asignatura in asignaturas:
    calificacion = input(f"Introduce tu nota en {asignatura}")
    notas.append(calificacion)
    

print("Vamos a mostrar ahora las notas:")
for i in range(len(asignaturas)):
    print(f"Asignatura: {asignaturas[i]} {notas[i]}")
```

```python
#otras alternativas, funcin zip
#para que el zip funcione bien ambas listas deben tener la misma longitud.
for asignatura, nota in zip(asignaturas, notas):
    print(f"Asignatura: {asignatura} nota: {nota}")
```

### Ejercicio 7

Nos han encargado que diseemos un programa para detectar bombas de una lista de N nmeros.

Sern bombas todos los nmeros que contengan la cifra que introduzca el jefe de los TEDAX por teclado(0-9)

**lista_con_bombas=[11, 107, 17, 67, 99, 45, 37, 87, 1007, 10007, 7, 1, 15, 81, 91, 88, 307]**

a) Analiza la lista para indicar el jefe TEDAX si hay bomba o la lista es segura.

b) Modifica el programa para en caso de existir bombas, indicar el n y la posicin en la que se encuentra de la lista de bombas


#### Solucin.

```python

lista_con_bombas = [11, 107, 17, 67, 99, 45, 37, 87, 1007, 10007, 7, 1, 15, 81, 91, 88, 307]
digito_bomba = int(input("Introduce el dgito para buscar bombas (0-9): "))

bombas = [num for num in lista_con_bombas if str(digito_bomba) in str(num)]

if bombas:
    print("Hay bombas en la lista!")
    for bomba in bombas:
        posicion = lista_con_bombas.index(bomba)
        print(f"Bomba encontrada: {bomba} en la posicin {posicion}")
else:
    print("La lista es segura, no hay bombas.")
```
