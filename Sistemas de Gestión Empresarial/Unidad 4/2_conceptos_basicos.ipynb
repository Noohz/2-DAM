### 1. Variables  y Tipos de Datos Primitivos

En programación, una variable está formada por un **espacio en la memoria del ordenador** y un **nombre simbólico** que está asociado a dicho espacio. Este espacio contiene una cantidad de información conocida o desconocida, es decir, **un valor o dato**  

---
  
Podemos imaginar una variable como una caja. En dicha caja podemos guardar cosas. Estas cosas son los datos. Para poder buscar las cajas(variables) en el almacén, les ponemos nombre.



---

En otros lenguajes, además del nombre, debemos decirle a la "caja" qué **tipo de dato** vamos a almacenar. En python esto **no es necesario**, recordemos que se trata de un lenguaje **dinamicamente tipado.**  

Veamos algunos ejemplos de definición de variables

```python
a = 5
print(type(a))
```

```python
cad = "cadena"
print(type(cad))
```

```python
real = 12.2
type(real)
```

```python
verdadero = True
type(verdadero)
```

```python
lista = [1, 'hola mundo', 9.7]
type(lista)
```

```python
conjunto = {1, 3.4, "adiós"}
type(conjunto)
```

### 5. Asignación Compacta

```python
i = 0
i = i +1
i
```

La asignación anterior es equivalente a:

```python
i = 0
i += 1
i
```

**Importante:** No puede haber espacio entre el + y el =  
Puedes incrementar una variable con cualquier cantidad, incluso con alguna que resulte de evaluar una expresión.

Todos los operadores aritméticos tienen su asignación con operadores asociada.

```python
i += 1
i = i +1
```

```python
i -= 1
i = i - 1
```

```python
i *= 2
i = i * 2
```

```python
i = 6

i = i / 2
print(i)
```
