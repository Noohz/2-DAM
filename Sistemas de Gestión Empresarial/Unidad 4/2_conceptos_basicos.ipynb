### 1. Variables  y Tipos de Datos Primitivos

En programacin, una variable est formada por un **espacio en la memoria del ordenador** y un **nombre simblico** que est asociado a dicho espacio. Este espacio contiene una cantidad de informacin conocida o desconocida, es decir, **un valor o dato**  

---
  
Podemos imaginar una variable como una caja. En dicha caja podemos guardar cosas. Estas cosas son los datos. Para poder buscar las cajas(variables) en el almacn, les ponemos nombre.



---

En otros lenguajes, adems del nombre, debemos decirle a la "caja" qu **tipo de dato** vamos a almacenar. En python esto **no es necesario**, recordemos que se trata de un lenguaje **dinamicamente tipado.**
